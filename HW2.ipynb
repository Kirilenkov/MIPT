### Задание №1

#### Найти пересечение без дубликатов

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел две последовательности чисел (некоторые идентификаторы товаров/услуг, просмотренных клиентом 'A' и клиентом 'B' на платформе «Авито»).

Необходимо конвертировать каждый ввод пользователя в отдельный список, состоящий из целых чисел, а также составить новый список, состоящий из элементов пересечения двух исходных. Дубликаты необходимо удалить, а результирующий список — вывести на экран.

**Важно!** Если пересечение пустое, то на экран должен выводиться пустой список.

Если ввод пользователя невозможно конвертировать в список из целых чисел, возникает ошибка. Необходимо отлавливать эту ошибку и выводить на экран фразу `Некорректный ввод`.

Для поиска пересечений в списках напишите функцию `pure_intersection()`, которая будет возвращать список с уникальными элементами пересечения.


**Пример №1:**
```
Ввод:
108, 138, 42, 52, 14
109, 13, 52, 32, 42, 14, 109

Вывод:
[42, 52, 14]

```

**Пример №2:**
```
Ввод:
108, 138, 42, 52, 14
189, 202, 249, 24, 15, 3

Вывод:
[]

```

**Пример №3:**
```
Ввод:
108, 138, 42, 52, четырнадцать
189, 202, 249, 24, 15, 3

Вывод:
Некорректный ввод

```

**Подсказка:**

Для определения пересечения преобразуйте списки во множества и воспользуйтесь методом `intersection()`.

**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**2 балла**):
    - каждый ввод пользователя конвертируется в список из идентификаторов;
    - функция для определения пересечения корректно работает для любого ввода пользователя, обозначенного в условиях задачи;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**1 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для определения пересечения работает только для ограниченного множества входных данных в рамках условия задачи;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм определения пересечения элементов.

````python
from functools import reduce


def lines2lists() -> list | None:

    """Функция line2lists вызывает два пользовательских ввода.
    Функция возвращает список из двух списков целых чисел типа int.
    Если ввод пользователя невозможно конвертировать в список
    из целых чисел, то на экран выводится фраза "Некорректный ввод",
    функция завершает работу и возвращает None."""

    try:
        output_list = []
        # Цикл сокращает напрасный труд пользователя, если некорректный ввод
        # был в первой последовательности, то вторую вводить не придется +
        # соблюдается принцип DRY
        for i in (1, 2):
            input_string = input(f'Введите {i}-ую последовательность идентификаторов: ')
            # Строка преобразуется в список строковым
            # методом split() с аргументом ‘, ’
            str_list = input_string.split(', ')
            # Вместо цикла для преобразования строковых значений списка
            # в целочисленные используется встроенная функция map
            output_list.append(list(map(int, str_list)))
    except ValueError:
        print('Некорректный ввод')
        return None
    else:
        return output_list


def pure_intersection(list1: list, list2: list, *other_lists: list) -> list:

    """Функция pure_intersection принимает два или более списков.
    Функция возвращает список с элементами,
    состоящий из элементов пересечения исходных списков."""

    # Формируется итерируемый объект типа map, преобразующий списки во множества
    set_map = map(set, [list1, list2] + list(other_lists))
    # С помощью функции reduce осуществляется вычисление итогового множества
    result_set = reduce(lambda prev_val, val: val.intersection(prev_val), set_map)
    # Возвращается итоговый список, полученный из итогового множества
    return list(result_set)


IDs = lines2lists()
# Если функция lines2lists отработала без исключений,
# то вызывается функция pure_intersection
if IDs is not None:
    print(pure_intersection(IDs[0], IDs[1]))
# Если я правильно понял, про удаление дубликатов:
del IDs
````

## Задание №2

#### Найти минимум и максимум

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь подаёт на вход программе действительные числа через пробел с клавиатуры (через функцию `input()`).

Он привык пользоваться Excel, поэтому отделяет дробную часть числа запятой, а не точкой. Кроме того, он может случайно ввести число «словами» (например, не "5", а «пять») — в этом случае слово надо игнорировать.

**Ваша задача** — вывести минимум и максимум последовательности. Гарантируется, что пользователем будет введено хотя бы одно число.

Для нахождения минимума и максимума напишите функцию `find_min_max`, которая будет возвращать два числа.

**Пример:**

```
Ввод:
-2,56 25,002 восемь 19,13 13 -37,5

Вывод:
Minimum: -37.5
Maximim: 25.002
```



**Подсказка №1:**

Для проверки того, состоит ли строка `s` только из букв, можно воспользоваться методом `s.isalpha()`.

**Подсказка №2:**

Для поиска минимума и максимума в списке можно воспользоваться встроенными функциями `min()` и `max()`.

**Критерии оценивания:**

- Задача решена корректно, учтены все условия (**2 балла**):
    - программа корректно обрабатывает ввод пользователя и работает на любых входных данных;
    - программа корректно обрабатывает присутствие слов во вводе пользователя;
    - программа корректно рассчитывает минимум и максимум числовой последовательности;
    - решение сопровождено комментариями.
- Задача решена корректно, но не выполнено одно из условий (**1 балл**):
    - программа некорректно обрабатывает ввод пользователя и работает только на конкретном примере входных данных;
    - программа некорректно обрабатывает присутствие слов во вводе пользователя, что приводит к ошибкам при выполнении программы;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): программа некорректно рассчитывает минимум и максимум числовой последовательности.

````python
input_string = input('Введите последовательность чисел: ')


def line2list(raw_srt: str) -> list:

    """Функция line2list принимает аргумент raw_srt типа str,
    В теле функции происходит преобразование полученной строки
    в список строк, каждое значение которого подвергается попытке
    преобразования в значение типа float. Функция возвращает
    список значений типа float, или пустой список."""

    elem_list = raw_srt.split(' ')
    result_list = []
    for elem in elem_list:
        try:
            result_list.append(float(elem.replace(',', '.')))
        except ValueError:
            pass
            # Если элемент невозможно преобразовать во float,
            # будь то слово или символ, то некорректный элемент
            # будет проигнорирован без сообщения пользователю и
            # экстренного завершения работы программы
            # В таком варианте покрывается больше кейсов некорректного ввода,
            # по сравнению с использованием метода .isalpha(), который ввернет
            # false, например, в случае применения к строкам 'восем,ь' или '!@#$%^'
    return result_list


def find_min_max(string: str) -> tuple:

    """Функция find_min_max принимает аргумент string типа str, далее отправляет
    его в функцию line2list, которая возвращает список значений типа float.
    После чего происходит поиск максимума и минимума значений списка
    стандартными списочными методами.  Функция возвращает кортеж из
    отформатированных строк или же кортеж из двух пустых строк,
    если line2list вернула пустой список"""

    target_list = line2list(string)
    # Если список для поиска минимум и максимума не пуст,
    # то возвращается кортеж из двух содержательных строк
    if target_list:
        return f'Minimum: {min(target_list)}\n', f'Maximum: {max(target_list)}'
    # Если список пуст, то возвращается две пустых строки
    else:
        return '', ''


# Для оценки работоспособности
print(*find_min_max(input_string), sep='')
````

#### Найти медиану

**Компания:** «Авито»

**Уровень:** Лёгкий

---
Медиана, как и среднее, является одной из важнейших характеристик для числовых данных в статистике.

**Медиана** — это число, которое находится в середине упорядоченной по возрастанию числовой последовательности.

Например, в последовательности чисел `1, 5, 2, 3, 6` медианой является число `3`, так как если упорядочить числа по возрастанию (`1, 2, 3, 5, 6`), центральным элементом в последовательности будет именно `3`.

**Примечание.** *Если в выборке чётное число элементов, то за медиану принимается среднее между центральными элементами (после упорядочивания).*

Например, в последовательности чисел `100, 5, 2, 4, 3, 6` медианой является число `4.5`, так как, если упорядочить числа по возрастанию (`2, 3, 4, 5, 6, 100`), то центральным элементом в последовательности будет среднее арифметическое между 4 и 5 ($\frac{4 + 5} {2} = 4.5$).

---
Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел неупорядоченную последовательность чисел.

Необходимо конвертировать ввод пользователя в список действительных чисел и вывести на экран медиану этой числовой последовательности.

Для поиска медианы в списке напишите функцию `find_median()`, которая будет возвращать одно число — медианное значение. Результат работы функции нужно вывести на экран.

Если переданная в функцию последовательность окажется пустой (пользователь ничего не ввёл) или пользователь ввёл числа некорректно (ввод нельзя преобразовать в числа), на экран должна быть выведена фраза `Некорректный ввод`.

**Пример №1:**

```

```
```
Ввод:
```

```
Вывод:
Некорректный ввод
```


**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**2 балла**):
    - ввод пользователя конвертируется в список из чисел;
    - функция для расчёта медианы работает корректно для списка чётной и нечётной длины;
    - в случае некорректного ввода функция возвращает `None`;
    - решение сопровождено комментариями.

    - ввод пользователя не конвертируется в список;
    - функция для расчёта медианы некорректно работает для списка чётной или нечётной длины;
    - в случае некорректного ввода функция отрабатывает с ошибкой;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм расчёта медианы для обоих случаев, результат работы программы зависит от вводимой последовательности чисел и т. д.

````python
# В формулировке к задаче нет ограничений на использование сторонних
# модулей. Имеется валидизированная продвинутая и почти наверняка более быстрая
# реализации расчета медианы, если сравнивать с самописной:
from statistics import median as md
input_string = input('Введите последовательность чисел: ')


def string2list(raw_srt: str) -> list | None:

  """Функция string2list принимает в качестве аргумента строку
  пользовательского ввода. В случае если пользовательский ввод
  был осуществлён корректно, функция возвращает список из натуральных
  чисел типа int. В ином случае функция выводит пользователю сообщение
  "Некорректный ввод" и возвращает None"""

  result_list = []
  for num in raw_srt.split(', '):
      # Помимо слов могут быть введены и другие артефакты, например,
      # лишние символы, которые не определяются методом isalpha()
      # Но их можно обработать с помощью try...except.
    try:
      result_list.append(int(num))
    except ValueError:
      # Функция print печатает 'Некорректный ввод' и возвращает объект None,
      # который в свою очередь возвращает функция strig2list
      return print('Некорректный ввод')
  # Если список не пустой, тогда идем по ветке if
  if result_list:
    return result_list
  # В ином случае идем по ветке else
  else:
    return print('Некорректный ввод')


def list_median(lst: list | None) -> float | None:

  """Функция list_median принимает в качестве аргумента список
  значений типа int. В теле функции осуществляется сортировка
  списка по возрастанию. После чего возвращается медианное
  значение отсортированного списка в формате float. В случае если
  вместо непустого списка в функцию передается объект типа None, тогда
  функция возвращет None"""

  if lst is not None:
      lst = sorted(lst)
      return float(md(lst))
  else:
    return None


def find_median(user_data: str) -> float | None:

  """Функция find_median принимает в качестве аргумента строку
  пользовательского ввода. Функция возвращает медианное
  значение в формате float от пользовательского ввода.
  В случае некорректного ввода функция возвращает None"""

  return list_median(string2list(user_data))


# Вывод на экран результата работы функции:
median = find_median(input_string)
print(f'{"Median: " + str(median) if median is not None else ""}')
````

#### Разложить факториал на простые множители

**Компания:** «СБЕР»

___

Пользователь с клавиатуры (через функцию `input()`) вводит некоторое целое число `n`.

Напишите функцию `decompose_factorial()`, которая будет принимать на вход введённое пользователем число `n`, а затем вычислять факториал этого числа и выполнять разложение факториала на простые множители.

---

Напомним, что **факториал числа $n$ ($n!$)** — это произведение всех натуральных чисел от $1$ до $n$ включительно.

Например, факториал числа $3$ будет равен: $3! = 1 * 2 * 3 = 6$.

---

**Простое число** — число, которое делится (нацело) только на себя и на 1. Например, $13$ — простое число, так как оно делится только на себя (на $13$) и на $1$, а $14$ не является простым числом, так как оно делится на $7$ и на $2$.

Подробнее о простых числах можно почитать [здесь](https://dzen.ru/a/XqFMe_VWclT8NTss).

---
**Простые множители** — это простые числа (не считая 1), на которые число делится без остатка.

Например, для числа $10$ простыми множителями являются числа $5$ и $2$. Для числа $28$ множителями являются числа $7$ и $4$, но простыми множителями будут числа $7$ и $2$, так как $4$ — это $2 * 2$, а число $28$ представляется в виде $28 = 7 * 2^2$.

----

В результате своей работы функция должна возвращать строку вида (приведён пример для `n=12`):

`12! = 2^10 * 3^5 * 5^2 * 7 * 11`

Здесь число слева от ^ — множитель, число справа от ^ — его степень. При этом если множитель встречается один раз, знак степени не выводится. Вывод множителей должен производиться в порядке их возрастания.

**Пример №1:**
```

Вывод:
5! = 2^3 * 3 * 5
```

**Пример №2:**

**Подсказка №2:**

Алгоритм нахождения простых множителей числа вы можете найти [здесь](https://foxford.ru/wiki/informatika/razlozhenie-chisla-na-mnozhiteli-v-python).

````python
from collections import defaultdict as dd
from decimal import Decimal as dm


def fac(n: int) -> int:

    """Функция fac принимает натуральное число типа int и возвращает
       его факториал также в формате int."""

    if n in (0, 1):
        return 1
    else:
        return fac(n-1) * n


def factorizer(num: int) -> dd:

    """Функция factorizer принимает натуральное число типа int,
       раскладывает его на простые множители, которые возвращаются в
       в формате defaultdict."""

    i = dm(2)
    factors_dict = dd(int)
    while i * i <= dm(num):
        while num % i == 0:
            factors_dict[int(i)] += 1
            num = num / i
        i = i + 1
    if num > 1:
        factors_dict[int(num)] = 1
    return factors_dict


number = int(input())
factorial = fac(number)
line4print = f'{number}! = '
if factorial == 1:
    line4print += '1'
else:
    prime_factors_dict = factorizer(factorial)
    line4print = f'{number}! = '
    list4print = []
    for k, v in prime_factors_dict.items():
        list4print.append(f'{k}^{v}' if v > 1 else f'{k}')
    line4print += ' * '.join(list4print)
print(line4print)
````

#### Заменить дублирующиеся буквы

**Компания:** «СБЕР»

____

Пользователь с клавиатуры (через функцию `input()`) вводит строку, состоящую из символов латинского алфавита.

Необходимо последовательно заменять две одинаковые буквы во введённой строке на следующую по алфавиту букву (важно: заменяем по 2 одинаковых буквы за одну итерацию, не более). Операция замены должна повторяться до тех пор, пока в строке не останутся только уникальные символы.

**Несколько важных моментов:**

* zz заменяем на a;
* вводимая строка состоит из букв латинского алфавита в нижнем регистре.

Напишите функцию `replace_duplicates()`, которая принимает на вход произвольную строку `s`, значение которой вводится с клавиатуры. Функция должна последовательно производить описанную операцию и возвращать строку с оставшимися буквами в любом порядке.

**Пример:**

**Подсказка:**

При решении задачи вам может понадобиться латинский алфавит. Вы можете получить его из встроенной в Python библиотеки `string`:

````python
import string
alphabet = string.ascii_lowercase
print(alphabet)
````

````python
from collections import defaultdict as dd
# Вместо строки с алфавитом можно использовать коды позиций в кодировке ASCII


def line_concat(ascii_dict: dict | dd) -> str:

    """Функция для конвертации словаря ASCII символов в строку."""

    line = ''
    for k, v in ascii_dict.items():
        if v > 0:
            line += chr(k)*v
    return line


def line2dd(line: str) -> dd:

    """Функция для конвертации строки в defaultdict словарь ASCII символов.
       Создается и возвращается словарь, где ключи - это коды ASCII символов,
       значения - количество повторений символа в строке"""

    char_dict = dd(int)
    for ch in line:
        char_dict[ord(ch)] += 1
    return char_dict


def replace_duplicates(line: str) -> str:

    """Функция replace_duplicates принимает строку, осуществляет проверку
       на повторяющиеся символы, производит замену одной найденной повторяющейся
       пары символов на один символ со следующим номером в таблице ASCII.
       Исключением является пара символов zz (ASCII код 122), они обрабатываются
       с заменой на символ a (ASCII код 97). Фунция возвращает строку с
       изменениями или без изменений, если таковых не требуется
       по условиям задачи."""

    char_dict = line2dd(line)
    tmp_ch = None
    for ch, n_ch in char_dict.items():
        if n_ch > 1:
            if ch != 122:
                char_dict[ch] -= 2
                tmp_ch = ch + 1
            else:
                char_dict[ch] -= 2
                tmp_ch = 97
            break
    if tmp_ch is not None:
        char_dict[tmp_ch] += 1
    return line_concat(char_dict)


ln = input('Введите строку: ')
tmp_dict = line2dd(ln)
flag = True
step = 0
# Основная управляющая конструкция, позволяющая последовательно
# вызывать функцию replace_duplicates и выводить в консоль результаты
# её работы
while flag:
    output_line = replace_duplicates(ln)
    output_dict = line2dd(output_line)
    if output_dict == tmp_dict:
        flag = False
        print('\noutput: \n', output_line, sep='')
    else:
        step += 1
        print(f'step {step}:', output_line)
    tmp_dict = output_dict
    ln = line_concat(output_dict)
````
